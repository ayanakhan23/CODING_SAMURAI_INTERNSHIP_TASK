```python
# Load the CSV file into a DataFrame
df = pd.read_csv(r"C:\Users\ayana\Downloads\amazon_csv_extract\amazon.csv")
```

```python
# Display the first few rows to verify the data is loaded correctly
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Converting Price Columns (discounted_price, actual_price)
# Remove ₹ and commas, then convert to float
df['discounted_price'] = df['discounted_price'].replace({'₹': '', ',': ''}, regex=True).astype(float)
df['actual_price'] = df['actual_price'].replace({'₹': '', ',': ''}, regex=True).astype(float)
```

```python
# Check the result
print(df[['discounted_price', 'actual_price']].head())
```

```python
#Converting Discount Percentage
# Remove % and convert to float
df['discount_percentage'] = df['discount_percentage'].replace({'%': ''}, regex=True).astype(float)
```

```python
# Check the result
print(df['discount_percentage'].head())
```

```python
# Display the unique values in the 'rating' column
print(df['rating'].unique())
```

```python
# Replace the invalid rating ('|') with NaN
df['rating'] = df['rating'].replace({'|': None})
```

```python
# Convert the ratings to numeric (any invalid data will be coerced to NaN)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
# Check for NaN values (after conversion)
print(f"Number of missing ratings: {df['rating'].isnull().sum()}")
```

```python
# Fill the missing rating with the mean of the 'rating' column
df['rating'] = df['rating'].fillna(df['rating'].mean())
```

```python
# Verify that there are no missing values now
print(f"Number of missing ratings after filling: {df['rating'].isnull().sum()}")
```

```python
# Display the cleaned ratings column
print(df[['rating']].head())
```

```python
# Inspect the unique values in 'rating_count' before converting to numeric
print(df['rating_count'].unique())
```

```python
# Remove commas and convert to numeric
df['rating_count'] = df['rating_count'].replace({',': ''}, regex=True).astype(float)
```

```python
# Fill NaN values with 0 (if necessary)
df['rating_count'] = df['rating_count'].fillna(0)
```

```python
# Verify the result
print(f"Number of missing values in 'rating_count' after cleaning: {df['rating_count'].isnull().sum()}")
print(df[['rating_count']].head())
```

```python
# Check data types after conversion
print(df.dtypes)
```

Sales by Category

```python
# Sum of rating_count by category
sales_by_category = df.groupby('category')['rating_count'].sum().sort_values(ascending=False)
```

```python
# Limit to top 10 categories for clarity
top_sales_by_category = sales_by_category.head(10)
```

```python
# Plot sales by category
plt.figure(figsize=(12, 6))
top_sales_by_category.plot(kind='bar', color='skyblue')


# Adding labels and title
plt.xlabel('Category')
plt.ylabel('Total Sales (rating_count)')
plt.title('Top 10 Sales by Product Category')


# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')



# Display the chart
plt.tight_layout()
plt.show()
```

Discount Percentage vs Rating

```python
# Plot a scatter plot of discount vs rating
plt.figure(figsize=(10, 6))
plt.scatter(df['discount_percentage'], df['rating'], alpha=0.5)
plt.title('Discount Percentage vs Rating')
plt.xlabel('Discount Percentage')
plt.ylabel('Product Rating')
plt.show()
```

Average Rating and Discount Analysis

```python
# Calculate average rating and discount percentage by product
avg_rating_discount = df.groupby('product_name')[['rating', 'discount_percentage']].mean()
```

```python
# Display the top 5 products based on average rating
print(avg_rating_discount.head())
```
